```python
# read in json file
yelp_df = pd.read_json('/Users/krystlelee/Documents/yelp_academic_dataset_business_revised.json')
```

```python
# view first row for reference of columns and value formats
yelp_df.head(1)
```

```python
# set index of dataset to business id
yelp_df_busid = yelp_df.set_index('business_id')
```

```python
# subset columns with non null values for categories column
cat_non_null = yelp_df_busid[yelp_df_busid['categories'].notnull()]
```

```python
# create a list of all categories from all rows where categories value is not null
# create empty list
cat_values = []
# iterate over the rows
for index, row in cat_non_null.iterrows():
    # add to list all categories by using extend method
    cat_values.extend(row['categories'].split(sep=','))
```

```python
# create list of unique values from categories column
output = []
# check each item in categories values
for x in cat_values:
    # remove white space from values
    clean_x = x.strip()
    # check for values, if not in list add otherwise do nothing
    if clean_x not in output:
        output.append(clean_x)
```
