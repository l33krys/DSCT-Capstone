```python
# create new data frame with business ids as index and attributes as columns
# create empty dictionary
diction = {}
# iterate over the rows
for index, row in yelp_df_busid.iterrows():
    # add keys, values to the dictionary using business id and attributes row
    diction[index] = row['attributes']
# create data frame of business id and attributes, transpose
business_att = pd.DataFrame(diction).transpose()   
```
